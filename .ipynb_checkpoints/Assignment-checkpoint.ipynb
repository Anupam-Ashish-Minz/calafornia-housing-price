```python
from langchain_community.llms import Ollama
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("housing.csv")
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.ocean_proximity.unique()
```

## graphs

### Histogram
The following graphs are histograms, they are used here to show the distribution of the data

```python
df.hist(figsize=(15, 8), bins=50)
```

### Heatmap
The following is a heatmap, it is used here to show the correlation of attributes with respect to each other

```python
ndf = df.select_dtypes(include=[np.number])
plt.figure(figsize=(15, 8))
sns.heatmap(ndf.corr(), annot=True)
```

## llm

As median_house_value is the only value that is of important to us, 
we are only generating the correlations with respect to that particual value

```python
corr = ndf.drop(['median_house_value'], axis=1).corrwith(df.median_house_value)
print(corr)
```

We are doing the same with the ocean proximity values, as the values are strings we have to use special functions

```python
ocorr = df.ocean_proximity.str.get_dummies().corrwith(df.median_house_value)
print(ocorr)
```

Here we are injecting the desciption of the data and the correlation values calculated above into the llm

```python
system_message = f"""
you are currently working on the calafornia housing price dataset
some key indication in the dataset are as follows
{df.describe()}
the correlation of median house values with respect to give paramters are as follows
{corr}
and the correlation of median house values with respect to proximity of ocean is as follows
{ocorr}
"""
```

```python
llm = Ollama(model="mistral", system=system_message)
for s in llm.stream("what are some characteristics of the give dataset"):
    print(s, end="")
```

```python
for s in llm.stream("which condition affect the median house price the most"):
    print(s, end="")
```
